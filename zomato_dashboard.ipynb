```python
# *******************************************************************
# Insights Dashboard

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("zomato_cleaned.csv")

plt.figure(figsize=(8,5))
plt.hist(df['rate'])
plt.xlabel("Rating")
plt.ylabel("Count")
plt.title("Rating Distribution")
plt.show()

# *******************************************************************

# *******************************************************************
# Dashboard made interactive

import plotly.express as px
import plotly.graph_objects as go

fig1 = px.histogram(df, x="rate", title="Rating Distribution (Interactive)")
fig1.show()

fig2 = px.box(df, y="rate", title="Rating Boxplot")
fig2.show()

numeric = df.select_dtypes(include=['number']).columns.tolist()

if len(numeric) >= 2:
    fig3 = px.scatter(df, x=numeric[0], y=numeric[1], color="rate",
                      title=f"Scatter Plot: {numeric[0]} vs {numeric[1]}")
    fig3.show()

# *******************************************************************
```
